```python
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style("grade3")
```

# Non-IID

```python
path = './dpclient'
client = '_0123456'
```

```python
client_out_dp = np.loadtxt(path+"_out"+"_dp"+client+".txt")
client_in_dp = np.loadtxt(path+"_in"+"_dp"+client+".txt")

client_out = np.loadtxt(path+"_out"+client+".txt")
client_in = np.loadtxt(path+"_in"+client+".txt")

xrange = np.arange(1, len(client_in_dp)+1)
```

```python
fig, (axe1, axe2) = plt.subplots(1, 2, figsize=(24,8))
axe1.plot(xrange, client_out_dp, 'g*--')
axe1.plot(xrange, client_in_dp, 'bv--')
axe1.legend(["out, %.3f"%(client_out_dp.mean()), "in, %.3f"%(client_in_dp.mean())])
axe1.set_title("loss=%.5f"%(np.abs(np.log(client_out_dp.mean()/client_in_dp.mean()))))
axe2.plot(xrange, client_out, 'g*--')
axe2.plot(xrange, client_in, 'bv--')
axe2.legend(["out, %.3f"%(client_out.mean()), "in, %.3f"%(client_in.mean())])
axe2.set_title("loss=%.5f"%(np.abs(np.log(client_out.mean()/client_in.mean()))))
plt.suptitle("(5,1e-4)-DP")
plt.show()
```
