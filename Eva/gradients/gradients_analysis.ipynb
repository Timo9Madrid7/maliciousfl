```python
import numpy as np

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from matplotlib import pyplot as plt

from jupyterthemes import jtplot
jtplot.style("grade3")
import warnings
warnings.filterwarnings('ignore')
```

```python
pca = PCA(n_components=2)
tsne = TSNE(n_components=2)
```

```python
grad0 = np.loadtxt('./gradients_0.txt')
grad1 = np.loadtxt('./gradients_1.txt')
```

```python
pca.fit(np.vstack((grad0, grad1)))
grad0_pca = pca.transform(grad0)
grad1_pca = pca.transform(grad1)

grad0_tsne = tsne.fit_transform(grad0)
grad1_tsne = tsne.fit_transform(grad1)
```

```python
plt.figure(figsize=(25,10))
plt.subplot(121)
plt.scatter(grad0_pca[:,0], grad0_pca[:,1], cmap='rainbow')
plt.scatter(grad1_pca[:,0], grad1_pca[:,1], cmap='rainbow')
plt.title('PCA')
plt.subplot(122)
plt.scatter(grad0_tsne[:,0], grad0_tsne[:,1], cmap='rainbow')
plt.scatter(grad1_tsne[:,0], grad1_tsne[:,1], cmap='rainbow')
plt.title("t-SNE")
plt.show()
```
